## Question 1
A permutation `perm` of `n + 1` integers of all the integers in the range `[0, n]` can be represented as a string `s` of length `n` where:
* `s[i] == 'I'` if `perm[i] < perm[i + 1]`, and
* `s[i] == 'D'` if `perm[i] > perm[i + 1]`.

Given a string `s`, reconstruct the permutation `perm` and return it. If there are multiple valid permutations `perm`, return any of them.

**Example:**

Input: `s = "IDID"`.

Output: `[0, 4, 1, 3, 2]`.

**Answer:**

```python
def perm_array(s: str):
    f = 0
    l = len(s)
    perm = []
    
    for char in s:
        if char == "I":
            perm.append(f)
            f += 1
        else:
            perm.append(l)
            l -= 1
    perm.append(f)
    
    return perm
```

```python
s = "IDID"

perm_array(s)
```

```python
s = "IDIIIDDIDDD"

perm_array(s)
```

## Question 2
You are given an `m × n` integer matrix `matrix` with the following two properties:
* Each row is sorted in non-decreasing order.
* The first integer of each row is greater than the last integer of the previous row.

Given an integer `target`, return `True` if `target` is in `matrix` or `False` otherwise. You must write a solution in $\mathcal{O}\left( \log{(mn)} \right)$ time complexity.

**Example:**



Input: `matrix = [[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]]`, `target = 3`.

Output: `True`.

**Answer:**

```python
def find_target_in_non_dec_matrix(matrix: list, target: int):
    n_rows = len(matrix)
    n_cols = len(matrix[0])

    start = 0
    end = n_rows*n_cols - 1

    while start <= end:
        mid = (start + end) // 2
        if matrix[mid // n_cols][mid % n_cols] == target:
            return True
        if matrix[mid // n_cols][mid % n_cols] < target:
            start = mid + 1
        else:
            end = mid - 1

    return False
```

```python
matrix = [
    [1,   3,  5,  7],
    [10, 11, 16, 20],
    [23, 30, 34, 60]
]

target = 3

find_target_in_non_dec_matrix(matrix, target)
```

## Question 3
Given an array of integers `arr`, return `True` if and only if it is a valid mountain array. Recall that `arr` is a mountain array if and only if:
* `arr.length >= 3`
* There exists some `i` with `0 < i < (arr.length - 1)` such that:
    - `arr[0] < arr[1] < ... < arr[i - 1] < arr[i]`
    - `arr[i] > arr[i + 1] > ... > arr[arr.length - 1]`

**Example:**



Input: `arr = [2, 1]`.

Output: `False`.

**Answer:**

```python
def mountain_array(arr: list):
    if len(arr) < 3:
        return False
    l = 0
    r = len(arr) - 1
    
    while l <= r and arr[l] < arr[l + 1]:
        l += 1
    
    while r >= 0 and arr[r] < arr[r - 1]:
        r -= 1
    return l == r
```

```python
mountain_array([2, 1])
```

```python
mountain_array([0, 2, 3, 4, 5, 2, 1, 0])
```

```python
mountain_array([0, 2, 3, 3, 5, 2, 1, 0])
```
